```python
### applies pre-processing steps to images
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
### binarization
# select threshold number

threshold = 172
source_dir = 'writable_area_notar_fr' 

def binarize_dir(threshold, dir):

    os.makedirs(f'preprocessed_imgs/binarized_imgs/{dir}', exist_ok=True)
        
    for file in os.listdir(dir):
        img_name = str(file)
        img = cv.imread(dir + '/' + img_name, cv.IMREAD_GRAYSCALE)
        retval, thresh = cv.threshold(img, threshold, 255, cv.THRESH_BINARY)
        cv.imwrite(f'preprocessed_imgs/binarized_imgs/{dir}/' + img_name, thresh)
        print(f'successfully wrote preprocessed_imgs/binarized_imgs/{dir}/' + img_name)

binarize_dir(threshold, source_dir)
```

```python
### visualize results
result_dir = os.listdir(f'preprocessed_imgs/binarized_imgs/{source_dir}')
plt.figure(figsize = (20, 20));
num = 16 # num of items to visualize
for i in range(1, num): 
    plt.subplot(4, 4, i); # number of rows, number of columns, item number
    img = mpimg.imread(f'preprocessed_imgs/binarized_imgs/{source_dir}/' + result_dir[i])
    plt.imshow(img, cmap = plt.cm.binary);
plt.show()
```
